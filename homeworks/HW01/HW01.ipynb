### 1) Импорт библиотеки NumPy с общепринятым сокращением.

```python
import numpy as np
```

### 2) Примеры создания массивов:

```python
# массив из обычного списка
nums_array = np.array([1, 2, 3]) 
nums_array
```

```python
# массив из нулей
zeros_array = np.zeros(8)
zeros_array
```

```python
# массив из едениц
ones_array = np.ones(10)
ones_array
```

```python
# массив из цифр (от 0 до 9)
nums_arange = np.arange(10)
nums_arange
```

### 3) Демонстрация атрибутов:

```python
# вывод формы (shape)
nums_arange.shape
```

```python
# вывод ипа данных (dtype)
nums_arange.dtype
```

### 4) Примеры индексирования и срезов:

```python
# получение отдельного элемента
nums_arange[4]
```

```python
# извлечение подмассива с помощью среза
nums_arange[2:7]
```

### 5) Примеры векторных операций и базовых статистик:

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python
# поэлементное сложение
a+b
```

```python
# поэлементное умножение массивов
a*b
```

```python
# вычисление средней величины (mean) 
(a+b).mean()
```
